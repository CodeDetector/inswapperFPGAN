```python
from swapper import process_video  , create_video 
import matplotlib.pyplot as plt 
model = "./checkpoints/inswapper_128_fp16.onnx"
video_path = "./Daawat World Biryani  Offline 60 Sec 19.05.24.mp4"
reference_images  = ["./Screenshot 2024-05-15 224011.png"] 
target_imgs= ["./Screenshot 2024-05-20 101516.png"]
result_image = process_video(video_path, target_imgs , reference_images , model  , restore = False )
```

```python
import cv2
import numpy as np 
frames = [] 

    
def create_video(new_frames ) : 
    
    h, w , _ = new_frames[0].shape
    fourcc = cv2.VideoWriter_fourcc(*'mpv4')
    fps = 30
    # size = (image_list[0].width, image_list[0].height)
    out = cv2.VideoWriter('output1.mp4', fourcc, fps , (w, h))

    # Write each image to the video
    for img in new_frames:
        out.write(img)

    # Release the video writer
    out.release()
create_video(np.array(result_image ))
```
